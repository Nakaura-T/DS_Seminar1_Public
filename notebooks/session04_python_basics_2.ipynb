## 3. Pandas：統計量を Pandas で理解する

データ分析の基本は「データの特徴を数値で要約すること」です。Pandasにはこれらの計算が最初から組み込まれています。

### 平均（Mean）

**全データの合計 ÷ 件数**。もっとも一般的な「代表値」です。ただし、飛び抜けて高い値（外れ値）があると引っ張られて実態から離れることがあります。


````python
mean_price = df["median_house_value"].mean()
print("平均価格:", mean_price)
````

### 中央値（Median）

データを小さい順に並べたとき、**ちょうど真ん中にある値**です。外れ値の影響を受けにくいという特徴があります。


````python
median_price = df["median_house_value"].median()
print("中央値（価格）:", median_price)
````

> [!NOTE]
> **平均 vs 中央値の違いを直感で理解する例**  
> 5人の年収が [300, 300, 350, 400, 5000（大富豪）] 万円の場合：  
> - 平均：(300+300+350+400+5000) / 5 = **1270万円**（大富豪に引っ張られて実態と乖離）  
> - 中央値：並べると真ん中は **350万円**（実態に近い）  
> 住宅価格など「外れ値が多いデータ」では中央値の方が実態を表すことが多いです。

### 標準偏差（Standard Deviation）

データが平均からどのくらい**ばらついているか**を示します。標準偏差が大きいほど、データがバラバラに散らばっています。


````python
std_price = df["median_house_value"].std()
print("標準偏差（価格）:", std_price)
````

### 四分位数（Quartile）

データを4等分したときの境界値です。外れ値を除いた「典型的な範囲」を把握するのに使います。

| 名称 | 記号 | 意味 | Pandasコード |
|---|---|---|---|
| 第1四分位数 | Q1 | 下位25%の境界 | `df["col"].quantile(0.25)` |
| 第2四分位数 | Q2 | 中央値（50%） | `df["col"].quantile(0.50)` |
| 第3四分位数 | Q3 | 上位25%の境界 | `df["col"].quantile(0.75)` |
| 四分位範囲 | IQR | Q3 - Q1（中央50%の幅） | `Q3 - Q1` |


````python
q1 = df["median_house_value"].quantile(0.25)
q2 = df["median_house_value"].quantile(0.50)
q3 = df["median_house_value"].quantile(0.75)
iqr = q3 - q1

print(f"Q1（25%点）: {q1:,.0f} ドル")
print(f"Q2（中央値）: {q2:,.0f} ドル")
print(f"Q3（75%点）: {q3:,.0f} ドル")
print(f"IQR（四分位範囲）: {iqr:,.0f} ドル")
````

### パラメトリック統計 vs ノンパラメトリック統計

統計的な分析手法には大きく2種類あります。どちらを使うかを決めるために、データの分布を確認する必要があります。

| | パラメトリック | ノンパラメトリック |
|---|---|---|
| **前提** | データが「正規分布（釣り鐘型）」に従う | 分布を仮定しない |
| **代表値** | **平均・標準偏差** を使う | **中央値・四分位数** を使う |
| **長所** | 統計的な検出力が高い | どんなデータにも使える |
| **例（検定）** | t検定・分散分析（ANOVA） | マン・ホイットニー検定・クラスカル・ウォリス検定 |
| **Pandasでの確認** | `.mean()`, `.std()` | `.median()`, `.quantile()` |

> [!NOTE]
> **どちらを使うか判断するには？**  
> ヒストグラムや箱ひげ図を見て、データが「釣り鐘型（左右対称）」に近ければパラメトリック、「偏り（skew）が強い」「外れ値が多い」ならノンパラメトリックを選ぶのが基本です。住宅価格データはどちらに見えますか？


````python
# 歪度（skewness）で確認することもできる（0に近いほど正規分布に近い）
print("歪度:", df["median_house_value"].skew())
````

## 7. Matplotlib：グラフの種類と「使い分け」

### Matplotlib とは？

**Matplotlib** は、Pythonで最も基本的なグラフ描画ライブラリです。「白紙のキャンバスに自分でグラフを描く」イメージで、全てを自分でコントロールできます。


````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # キャンバスのサイズ設定
# ↓ ここにグラフの種類を書く
plt.title("タイトル")
plt.xlabel("X軸の名前")
plt.ylabel("Y軸の名前")
plt.show()
````

### グラフの種類と使い分けガイド

データを見るとき、「**何を知りたいか**」によってグラフを選びます。

---

#### 1. ヒストグラム（`plt.hist`）：分布を見る

「どの価格帯の家が最も多いか？」のような、**1つのデータのばらつき・分布形状**を確認します。


````python
plt.figure(figsize=(10, 5))
plt.hist(df["median_house_value"], bins=50, color="steelblue", edgecolor="white")
plt.title("住宅価格の分布")
plt.xlabel("価格（ドル）")
plt.ylabel("件数")
plt.show()
````

🎯 **使いどき**：「このデータは釣り鐘型（正規分布）か？」「外れ値はあるか？」を確認したいとき

---

#### 2. 散布図（`plt.scatter`）：2変数の関係を見る

「収入が高い地域ほど家も高いか？」のような、**2つの変数の関係（相関）**を確認します。


````python
plt.figure(figsize=(10, 6))
plt.scatter(df["median_income"], df["median_house_value"],
            alpha=0.3, color="coral", s=10)
plt.title("収入と住宅価格の関係")
plt.xlabel("収入")
plt.ylabel("住宅価格")
plt.show()
````

🎯 **使いどき**：「2つの変数に相関（関係）があるか？」を確認したいとき

---

#### 3. 折れ線グラフ（`plt.plot`）：推移・傾向を見る

「築年数が増えると価格はどう変化するか？」のような、**時系列や順序があるデータの推移**を表します。


````python
# 築年数ごとの平均価格を折れ線で表示
age_price = df.groupby("housing_median_age")["median_house_value"].mean()

plt.figure(figsize=(12, 5))
plt.plot(age_price.index, age_price.values,
         color="navy", linewidth=2, marker="o", markersize=4)
plt.title("築年数ごとの平均住宅価格")
plt.xlabel("築年数")
plt.ylabel("平均価格")
plt.grid(True)
plt.show()
````

🎯 **使いどき**：「時間の流れ」や「連続した変化」を見たいとき

---

#### 4. 棒グラフ（`plt.bar`）：カテゴリ間を比較する

「グループAとグループBでどちらが多いか？」のような、**カテゴリ間の大きさの比較**に使います。


````python
# 築年数グループごとの件数を棒グラフで表示
age_groups = df["housing_median_age"].value_counts().sort_index()

plt.figure(figsize=(14, 5))
plt.bar(age_groups.index, age_groups.values, color="mediumseagreen")
plt.title("築年数ごとの件数")
plt.xlabel("築年数（年）")
plt.ylabel("件数")
plt.show()
````

🎯 **使いどき**：「カテゴリの大小を比べたい」とき

#### 5. 箱ひげ図（`plt.boxplot`）：分布の要約を見る

平均だけでなく、**中央値・四分位数・外れ値**を1つのグラフで確認できます。


````python
plt.figure(figsize=(8, 5))
plt.boxplot(df["median_house_value"], vert=True)
plt.title("住宅価格の箱ひげ図")
plt.ylabel("住宅価格（ドル）")
plt.show()
````

**箱ひげ図の読み方：**
```
 ─┬─        ← 上限（外れ値除く最大）
  │
  ┌─┐       ← Q3（75%点）
  │ │  ← この箱の中に全データの50%が入る
  ├─┤       ← Q2（中央値）
  │ │
  └─┘       ← Q1（25%点）
  │
 ─┴─        ← 下限（外れ値除く最小）
  ○  ○     ← 外れ値（異常に高い・低い値）
```

🎯 **使いどき**：「データのばらつきと外れ値」を同時に確認したいとき。ノンパラメトリックな統計（中央値・四分位数）で表現したいとき

---

#### 6. 円グラフ（`plt.pie`）：割合を見る

全体に対する各カテゴリの**比率・構成比**を表します。


````python
# 築年数を3グループに分けて割合を表示
bins = [0, 20, 40, 100]
labels = ["新しい（0-20年）", "中程度（21-40年）", "古い（41年〜）"]
df["age_cat"] = pd.cut(df["housing_median_age"], bins=bins, labels=labels)

counts = df["age_cat"].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
plt.title("築年数の割合")
plt.show()
````

🎯 **使いどき**：「全体の中でどのカテゴリが何%か」を表したいとき（3〜5種類程度が見やすい）

## 9. Seaborn：きれいなグラフを手軽に描く

### Seaborn とは？

**Seaborn** はMatplotlibをベースに作られた、より美しく・より簡単にグラフを描けるライブラリです。

- **Matplotlib**：「白紙から全て自分で描く」→ 自由だが手間がかかる
- **Seaborn**：「テンプレートをベースに描く」→ 少ないコードで見栄えが良い


````python
import seaborn as sns

sns.set_theme(style="whitegrid")  # テーマ設定（whitegrid, darkgrid, white, dark, ticks）
````

### histplot：KDE 曲線付きヒストグラム


````python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="median_house_value", bins=50,
             kde=True,          # KDE曲線（分布の形を滑らかに）を重ねる
             color="steelblue")
plt.title("住宅価格の分布（KDE付き）")
plt.show()
````

### scatterplot：自動的に見やすい散布図


````python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="median_income", y="median_house_value", alpha=0.3)
plt.title("収入と住宅価格")
plt.show()
````

### boxplot：グループ別の分布比較


````python
df["age_group"] = (df["housing_median_age"] // 10 * 10).astype(int)

plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x="age_group", y="median_house_value", palette="Blues")
plt.title("築年数グループごとの住宅価格分布")
plt.show()
````

### 📝 演習 9-1

`sns.histplot()` で `median_income`（収入）のヒストグラムを `kde=True` で描き、Matplotlibのヒストグラムと見た目を比べてください。

## 10. Seaborn：高度な可視化

### heatmap：相関行列で関係を一目で見る

**相関係数**は、2つの変数が「どのくらい一緒に動くか」を -1 〜 +1 の数値で表します。

| 値の範囲 | 意味 |
|---|---|
| +0.7 〜 +1.0 | 強い正の相関（一方が増えると他方も増える） |
| +0.3 〜 +0.7 | 中程度の正の相関 |
| -0.3 〜 +0.3 | ほとんど関係がない |
| -0.7 〜 -0.3 | 中程度の負の相関 |
| -1.0 〜 -0.7 | 強い負の相関（一方が増えると他方は減る） |


````python
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,       # 各セルに数値を表示
    fmt=".2f",        # 小数2桁
    cmap="coolwarm",  # 赤（正）〜青（負）のカラーマップ
    vmin=-1, vmax=1,
    linewidths=0.5
)
plt.title("住宅データの相関行列")
plt.show()
````

### pairplot：全変数の関係を一度に見る


````python
# データを500件にサンプリング（全件だと時間がかかる）
df_sample = df.sample(500, random_state=42)

sns.pairplot(
    df_sample[["median_income", "median_house_value", "housing_median_age"]],
    alpha=0.3
)
plt.show()
````

````python
# ステップ1：データを読み込み、基本統計を確認する
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
print("件数:", df.shape)
df.describe()
````

````python
# ステップ2：価格の平均・中央値・標準偏差・四分位数をすべて表示する
print("平均:", df["median_house_value"].mean())
print("中央値:", df["median_house_value"].median())
print("標準偏差:", df["median_house_value"].std())
print("Q1:", df["median_house_value"].quantile(0.25))
print("Q3:", df["median_house_value"].quantile(0.75))
````

````python
# ステップ3：高価格帯（Q3以上）と低価格帯（Q1以下）を絞り込んで件数を比較する
q1 = df["median_house_value"].quantile(0.25)
q3 = df["median_house_value"].quantile(0.75)

expensive = df[df["median_house_value"] >= q3]
cheap     = df[df["median_house_value"] <= q1]
print("高価格帯:", expensive.shape[0], "件")
print("低価格帯:", cheap.shape[0], "件")
````

````python
# ステップ4：2行2列のグラフ（価格ヒストグラム・収入ヒストグラム・散布図・相関ヒートマップ）を描く
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].hist(df["median_house_value"], bins=40, color="steelblue")
axes[0, 0].set_title("住宅価格の分布")

axes[0, 1].hist(df["median_income"], bins=40, color="coral")
axes[0, 1].set_title("収入の分布")

axes[1, 0].scatter(df["median_income"], df["median_house_value"], alpha=0.2, s=5)
axes[1, 0].set_title("収入と価格の関係")

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", ax=axes[1, 1])
axes[1, 1].set_title("相関行列")

plt.tight_layout()
plt.show()
````
