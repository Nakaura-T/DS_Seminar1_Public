## 5. 比較演算子：「= 代入」と「== 比較」の違い

プログラムで最も間違えやすいポイントです。

- **`=`（イコール1つ）**：変数に値を**代入（保存）**する
- **`==`（イコール2つ）**：左右の値が**等しいか比較**する（True/Falseを返す）


````python
score = 90           # 代入：scoreに90を保存する

print(score == 90)   # 比較：scoreは90と等しいか？ → True
print(score == 80)   # 比較：scoreは80と等しいか？ → False
````

その他の比較演算子：

| 記号 | 意味 | 例 | 結果 |
|---|---|---|---|
| `==` | 等しい | `5 == 5` | `True` |
| `!=` | 等しくない | `5 != 3` | `True` |
| `>` | より大きい | `10 > 5` | `True` |
| `<` | より小さい | `3 < 1` | `False` |
| `>=` | 以上（自分も含む） | `5 >= 5` | `True` |
| `<=` | 以下 | `4 <= 3` | `False` |

## 6. 変数の更新（`+=`）

変数の値を少しずつ変えていく場合、次のような**更新の書き方**をよく使います。


````python
count = 0
print(count)   # 0

count = count + 1   # count に 1 を足して、また count に保存しなおす
print(count)   # 1

# 上の書き方を短く書ける「+=」
count += 1
print(count)   # 2

# -= でも同様
score = 100
score -= 10
print(score)   # 90
````

## 7. リスト：データの並び（プレイリスト）

1つの変数に、たくさんのデータを順番に並べて保存したいときは**リスト**を使います。  
音楽の「プレイリスト」やLINEの「友達リスト」と同じです。

- **カッコ `[ ]`** で囲み、中身を **カンマ `,`** で区切ります。
- **0番目から数える**：1番ではなく0番から数えるルールに注意してください。


````python
# お気に入りのフルーツリスト
fruits = ["りんご", "みかん", "スイカ", "メロン"]

print(fruits[0])   # 0番目 → 「りんご」
print(fruits[2])   # 2番目 → 「スイカ」
print(fruits[-1])  # 後ろから1番目 → 「メロン」

# リストの長さ（個数）を調べる
print(len(fruits))  # 4

# リストに要素を追加する
fruits.append("ぶどう")
print(fruits)
````

## 8. Pythonの最重要ルール：インデント（段落）

ここから「条件によって動きを変える（if）」や「繰り返し（for）」を学びますが、その前にPythonで最も重要なルールを説明します。

プログラミングでは「ここからここまでが命令のまとまり（ブロック）です」と指定する必要があります。  
多くの言語では `{ }`（波カッコ）を使いますが、**Pythonではインデント（行の前の空白）**を使って表現します。

```
他の言語（C言語やJavaのイメージ）:        Python:
if (score >= 60) {                         if score >= 60:
    print("合格");                             print("合格")
}
```

**ルールのまとめ：**

1. `if` や `for` の行の末尾に **コロン `:`** を書く → 「ここから命令のまとまりが始まる」合図
2. `:`の次の行からは **Tab キー**（または半角スペース4つ）で段落を一段下げる
3. 段落を元の位置に戻すと、そのまとまりは終わる

> [!CAUTION]
> **「全角スペース」は絶対に禁止です！**  
> 見た目には区別がつきませんが、全角スペースが混ざると `IndentationError` で止まります。

## 9. もし〜なら（if 文）

条件によって動きを変えます。「バッテリーが少なくなったら省電力モード」などのスマホの設定と全く同じ仕組みです。


````python
battery = 15

if battery <= 20:
    # インデントが下がっている = if が True のときだけ実行される
    print("バッテリーが少ないです！")
    print("省電力モードにします。")
elif battery <= 50:
    # elif = else if（さらに別の条件を追加できる）
    print("バッテリーが半分以下です。")
else:
    # 上の条件がどれも当てはまらなかった場合
    print("バッテリーは十分です。")

print("処理が終わりました。")  # インデントがないので必ず実行される
````

### `and` / `or`：複数の条件を組み合わせる


````python
age = 25
is_member = True

# and : 両方の条件が True のとき
if age >= 20 and is_member:
    print("会員割引が適用されます")

# or : どちらか一方が True のとき
score = 75
if score >= 90 or is_member:
    print("特典があります")
````

## 10. 繰り返し（for 文）

リストの中身を1つずつ取り出して、同じ処理を繰り返します。  
「LINEのグループ全員に同じメッセージを送る」自動化イメージです。


````python
members = ["田中", "佐藤", "鈴木"]

for person in members:
    # membersから1人ずつ取り出して「person」に入れる
    print(person + "さん、こんにちは！")

print("全員に送信完了")
````

### `range()`：数値の繰り返し

リストがなくても、数字の連番で繰り返せます。


````python
# range(5) は 0, 1, 2, 3, 4 を順に生成する（5は含まれない）
for i in range(5):
    print(i)

# range(1, 6) は 1, 2, 3, 4, 5 を生成する
for i in range(1, 6):
    print(i)
````

### `+=` と `for` の組み合わせ（合計を計算する）


````python
scores = [80, 90, 70, 85, 60]

total = 0           # 最初はゼロ
for s in scores:
    total += s      # 1つずつ足していく

print("合計:", total)
print("平均:", total / len(scores))
````

## 11. 繰り返し（while 文）

`for` が「リストの要素を全部処理したら終わり」なのに対して、`while` は**条件が満たされている間、ずっと繰り返す**ものです。

「充電が100%になるまでずっと充電し続ける」のようなイメージです。


````python
battery = 10

while battery < 100:
    print("充電中... 現在" + str(battery) + "%")
    battery += 10   # ← これがないと永遠に終わらない（無限ループ）に注意！

print("充電完了！")
````

> [!CAUTION]
> `while` の中で変数を更新しないと、条件が永遠に True のままとなり**無限ループ**になります。  
> その場合はセルの左の **■（停止ボタン）** をクリックして止めてください。

## 12. 関数を自分で作る（def）

何度も使う処理をひとまとめにして名前をつけたものを**関数（かんすう）**と呼びます。  
`def`（Define：定義する）という言葉を使って、「新しい命令」を自分で作ることができます。


````python
# 挨拶する関数「greet」を定義する
def greet(name):          # name が「引数（ひきすう）」= 外から渡す値
    print(name + "さん、こんにちは。")
    print("Pythonの学習を始めましょう！")

# 作った関数を呼び出す
greet("田中")   # "田中" を name に渡す
greet("佐藤")
````

### 戻り値（return）：計算結果を返す


````python
# BMIを計算して結果を返す関数
def calc_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi   # 計算結果を「呼び出し元」に返す

# 関数を使って計算する
my_bmi = calc_bmi(65, 1.70)
print(my_bmi)
````

## 13. ライブラリと `import`：便利な「道具箱」を借りる

Pythonには最初から組み込まれている命令のほかに、**ライブラリ**という「外から借りてくる道具箱」があります。

### なぜ `import` が必要なのか？

料理に例えると、Pythonそのものは「台所（基本的な環境）」です。  
ライブラリは「道具箱（包丁セット、フードプロセッサーなど）」で、普段は戸棚に仕舞われています。  
使う前に **「この道具箱を持ってきて！」** と指示するのが `import` です。


````python
# mathライブラリ（数学の道具箱）を使えるようにする
import math

print(math.sqrt(25))   # 平方根 → 5.0
print(math.pi)         # 円周率 → 3.14159...
````

### `import ... as ...`：あだ名をつけて短くする

毎回 `pandas.read_csv()` と書くのは長いので、`pd` という短いあだ名（エイリアス）をつけるのが慣習です。


````python
import pandas as pd              # pandas を「pd」というあだ名で使う
import matplotlib.pyplot as plt  # matplotlib.pyplot を「plt」というあだ名で使う
````

### よく使うライブラリ一覧

| ライブラリ | 通称 | 用途 |
|---|---|---|
| `pandas` | `pd` | 表データの操作・読み込み |
| `matplotlib.pyplot` | `plt` | グラフの描画 |
| `seaborn` | `sns` | きれいなグラフの描画 |
| `numpy` | `np` | 数値計算・行列演算 |
| `math` | （そのまま） | 数学関数（平方根・円周率など） |

## 14. Pandas：データの「表」を読み込む

データ分析で最も使われるライブラリが **Pandas（パンダス）** です。  
これを使うと、CSVなどの「表」を一瞬でPythonに取り込んで操作できます。


````python
import pandas as pd

# Google Colabに最初から用意されているデータを読み込む
# 引数に「ファイルの場所（パス）」を文字列で渡す
df = pd.read_csv("/content/sample_data/california_housing_train.csv")

# 最初の5行だけ表示
df.head()
````

````python
# データの基本情報を確認する
print(df.shape)    # 行数・列数
print(df.columns)  # 列名の一覧
df.describe()      # 平均・最大・最小などの統計情報
````
