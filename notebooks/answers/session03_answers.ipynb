````python
x = 60

print(x >= 60)  # True  （60以上 → 60自身も含む）
print(x > 60)   # False （60より大きい → 60は含まない）
````

## 演習 9-2：気温と天気による条件分岐


````python
temp = 28
is_sunny = True

if temp >= 30 and is_sunny:
    print("海に行こう！")
elif temp >= 30 or is_sunny:
    print("外出もいいかも")
else:
    print("家でゆっくりしよう")
````

## 演習 10-2：for と += で合計・平均を計算


````python
scores = [80, 90, 70, 85, 60]

total = 0
for s in scores:
    total += s

average = total / len(scores)
print("合計:", total)    # 385
print("平均:", average)  # 77.0
````

## 演習 11-1：n を2倍にし続けるwhile


````python
n = 1
while n <= 10:
    print(n)
    n *= 2  # n = n * 2 と同じ（n を2倍にする）

print("終了。最終値:", n)
````

**出力：**
```
1
2
4
8
終了。最終値: 16
```
ループ内で `print(n)` が実行されるのは `n <= 10` の間なので、表示されるのは **1・2・4・8** の4つです。`n = 8` の後に `n *= 2` で `n = 16` になった時点で条件が `False` になりループを抜け、最後の `print("終了。最終値:", n)` で16が表示されます（16はループ内では表示されません）。

## 演習 11-2：1から100までの合計


````python
total = 0
i = 1
while i <= 100:
    total += i
    i += 1

print("合計:", total)  # 5050
````

**答え：** 5050（ガウスの公式 100×101÷2 の答えと一致します）
