### 問題1: データの確認
`iris` データセットの最初の 10 行を表示し、データの行数と列数を表示してください。


```python
import seaborn as sns
iris = sns.load_dataset('iris')

# 最初の10行
print(iris.head(10))

# 行数と列数
print(f"Shape: {iris.shape}") # (150, 5)
```

---

### 問題2: 特定の列の抽出
`sepal_length` と `species` の 2 列だけを抽出した `df_sub` を作成してください。


```python
df_sub = iris[['sepal_length', 'species']]
print(df_sub.head())
```

---

### 問題3: 条件によるフィルタリング
`petal_length` が 5.0 以上の行だけを抽出して表示してください。


```python
long_petals = iris[iris['petal_length'] >= 5.0]
print(long_petals.head())
print(f"該当件数: {len(long_petals)}")
```

---

### 問題4: 代表値の計算
`petal_width` の「平均値」と「中央値」を計算して表示してください。


```python
mean_val = iris['petal_width'].mean()
median_val = iris['petal_width'].median()

print(f"平均値: {mean_val:.2f}")
print(f"中央値: {median_val:.2f}")
```

---

### 問題5: 要約統計量の一括表示
`iris` データセット全体の要約統計量を表示してください。


```python
print(iris.describe())
```

---

### 問題6: 分布の可視化
`sepal_width` のヒストグラムを作成してください。


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seabornのテーマを適用
sns.set_theme(style="whitegrid")

sns.histplot(iris['sepal_width'], kde=True, color="skyblue")
plt.title('Distribution of Sepal Width')
plt.show()
```

---

### 問題7: 種類ごとの比較（箱ひげ図）
アヤメの種類（`species`）ごとに、`petal_length` の箱ひげ図を作成してください。


```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_length', data=iris, palette="muted")
plt.title('Petal Length by Species')
plt.show()
```

---

### 問題8: 2変数の関係（散布図）
`sepal_length` (x) vs `petal_length` (y) の散布図と回帰直線を表示してください。


```python
plt.figure(figsize=(8, 5))
sns.regplot(x='sepal_length', y='petal_length', data=iris, line_kws={'color': 'darkred'})
plt.title('Sepal Length vs Petal Length with Regression Line')
plt.show()
```

---

### 問題9: 推移の可視化（ラインプロット）
インデックスを時間軸に見立てて、`sepal_length` の変化を表示してください。


```python
plt.figure(figsize=(10, 4))
sns.lineplot(data=iris['sepal_length'], color="teal", linewidth=1.5)
plt.title('Sepal Length Trend')
plt.show()
```
