## 1. セットアップ

世界中の学習済み AI が集まるプラットフォーム、Hugging Face の `transformers` ライブラリを使用します。

```python
# ライブラリのセットアップ（Colab の場合）
!pip install -q transformers sentencepiece torch

import torch
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification
import pandas as pd

print("Global Language Processing Engine, Initialized.")
```

## 2. 【講義1+2】言葉を数にする技術：Transformer の衝撃

### 2.1 単語のベクトル空間
かつて AI は「王様 - 男性 + 女性 = 女王」という計算を可能にしました (Word2Vec)。しかし、これでは「私は林檎を食べる」の林檎も「Apple 社の株価」の Apple も同じベクトルになってしまいます。これを解決したのが **文脈に応じたベクトル化** です。

### 2.2 Self-Attention：文の中の関連性
「彼女は林檎を食べた。それは甘かった。」という文において、AI は「それ」が「林檎」に近い関係であることを、行列演算による **アテンション（注意）** で計算します。これにより、長い文章の端と端にある情報の繋がりを捉えられるようになりました。

---

## 3. 【実習1】トークナイゼーションの理解

AI は「人間が思う単語」ではなく「トークン」という単位で言葉を処理します。

```python
# BERT のトークナイザをロード
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Immunotherapy is a breakthrough in oncology."
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print(f"Original Text: {text}")
print(f"Tokens: {tokens}") # 'oncology' が分割される場合があることに注目！
print(f"Token IDs: {token_ids}")
```

## 4. 【実習2】実務で使えるパイプライン操作

数行のコードで、世界最高水準の AI 機能を呼び出します。

```python
# 1. 感情分析 (Contextual Awareness)
sentiment_analyzer = pipeline("sentiment-analysis")
results = sentiment_analyzer([
    "This new treatment shows promising results for lung cancer.",
    "The generic drug failed to meet the required safety standards."
])
display(pd.DataFrame(results))

# 2. 固有表現抽出 (NER: Named Entity Recognition)
# 文中から「組織名」「地名」「人名」等を抜き出します
ner_engine = pipeline("ner", grouped_entities=True)
sample_text = "Dr. Kumamoto is working at the University Hospital in Japan since 2026."
ner_results = ner_engine(sample_text)
print("\n--- NER Results ---")
display(pd.DataFrame(ner_results))
```

## 5. 【講義3】BERT $vs$ GPT：2 つの巨大な山脈

- **BERT (Encoder系)**: 文の前後をじっくり読み、意味を深く理解する。穴埋め問題が得意。「この単語は、この文脈でどういう意味か？」の分類タスクに強い。
- **GPT (Decoder系)**: 文の続きを予測する能力に特化している。クリエイティブな文章生成や対話が得意。
